```python
import pandas as pd
input = ""
with open("./input.txt", 'r') as f:
    input = f.read().splitlines()

input
```

```python
xs = []
ys = []

for l in input:
    line = l.split()
    xs.append(line[0])
    ys.append(line[1])

for i, x in enumerate(xs):
    xs[i] = int(x)

for i, y in enumerate(ys):
    ys[i] = int(y)

xs.sort()
ys.sort()

data = pd.DataFrame(zip(xs, ys))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(xs, ys, marker='o')
plt.title('Data Visualization')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```

```python
data['loss'] = abs(data[1] - data[0])
```

```python
counts = {}
for row in data.itertuples(index=False):
    x, y = row.xs, row.ys
    if x not in counts:
        counts[x] = 0
    if y == x:
        counts[x] += 1
```
